```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# create a column age_group
labels=['young adult','adult','Middle_age','senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#purchase frequency days
frequency_mapping={
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365.0,
    'Every 3 Months':90.0
}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
import pandas as pd
from sqlalchemy import create_engine
import urllib
```

```python
server_name = r'SABI-21\SQLEXPRESS'      # Or your specific SQL Server instance name
database_name = 'customer_behaviour'
```

```python
params = urllib.parse.quote_plus(
    f'DRIVER={{ODBC Driver 17 for SQL Server}};'
    f'SERVER={server_name};'
    f'DATABASE={database_name};'
    f'Trusted_Connection=yes;'
)
```

```python
engine = create_engine(f"mssql+pyodbc:///?odbc_connect={params}")
```

```python
table_name = 'customer_shopping'    # What you want to name your new SQL table

df.to_sql(
    name=table_name, 
    con=engine, 
    if_exists='replace',            # Options: 'fail', 'replace', 'append'
    index=False                     # Set to False so Pandas doesn't turn the index into a column
)

print(f"Success! DataFrame loaded into table '{table_name}'.")
```
